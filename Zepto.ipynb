```python
df_clean["DiscountAmount"] = (
    df_clean["OriginalPrice"] - df_clean["Price"]
)
```

```python
df_clean[["Price", "OriginalPrice", "DiscountAmount"]].head()
```

```python
df_clean["DiscountPercent"] = (
    (df_clean["DiscountAmount"] / df_clean["OriginalPrice"]) * 100
).round(2)
```

```python
df_clean.columns
```

```python
#Price Category

df_clean["PriceCategory"] = pd.cut(
    df_clean["Price"],
    bins=[0,100,500,float("inf")],
    labels=["Budget","Mid Range","Premium"]
)
```

```python
df_clean["PriceCategory"].value_counts()
```

```python
#Rating Category

df_clean["RatingCategory"]=pd.cut(
    df_clean["Ratings"],
    bins=[0,4.0,4.5,5.0],
    labels=["Average","Good","Execellent"],
    include_lowest=True
)
```

```python
df_clean["RatingCategory"].value_counts()
```

```python
#Is Discount Available

df_clean["DiscountAvailable"] = np.where(
    df_clean["DiscountAmount"] > 0,
    "Yes",
    "No"
)
```

```python
df_clean["DiscountAvailable"].value_counts()
```

```python
#High Rated Product

df_clean["HighRated"] = np.where(
    df_clean["Ratings"] >= 4.5,
    "Yes",
    "No"
)
```

```python
df_clean["HighRated"].value_counts()
```

```python
#Premium Product

df_clean["PremiumProduct"] = np .where(
    df_clean["Price"] >= 1000,
    "Yes",
    "No"
)
```

```python
#Popularity Score

df_clean["PopularityScore"] = (
    df_clean["Ratings"] * np.log1p(df_clean["Review"])
).round(2)
```

```python
#Top 10 Most Popular Products

df_clean.nlargest(
    10,
    "PopularityScore"
)[["Name","Ratings","Review","PopularityScore"]]
```

```python
#Save the Enhabced Dataset

df_clean. to_csv(
    "zepto_feature_engineered.csv",
    index=False
)
```
